```python
import pandas as pd

# Load data
data = pd.read_csv("dna_profiles.csv")

# Extract profiles
def get_profile(profile_id):
    profile_data = data[data["id"] == profile_id]
    profile = {}

    for _, row in profile_data.iterrows():
        profile[row["locus"]] = (row["allele1"], row["allele2"])

    return profile

# Compare
def compare_profiles(p1, p2):
    match_count = 0
    mismatches = []

    for locus in p1:
        if locus in p2:
            if p1[locus] == p2[locus]:
                match_count += 1
            else:
                mismatches.append(locus)

    total_loci = len(p1)
    match_percentage = (match_count / total_loci) * 100

    return match_percentage, match_count, mismatches

# Run
profile1 = get_profile(1)
profile2 = get_profile(2)

result, matches, mismatches = compare_profiles(profile1, profile2)

print("FORENSIC DNA REPORT")
print("----------------------")
print("Matching Loci:", matches)
print("Mismatched Loci:", mismatches)
print(f"Match Percentage: {result:.2f}%")

# Save report
with open("report.txt", "w") as file:
    file.write("FORENSIC DNA ANALYSIS REPORT\n")
    file.write("-----------------------------\n")
    file.write(f"Matching Loci: {matches}\n")
    file.write(f"Mismatched Loci: {mismatches}\n")
    file.write(f"Match Percentage: {result:.2f}%\n")

print("Report saved as report.txt")
```
